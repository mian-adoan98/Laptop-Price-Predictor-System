```python
# Load dataset
data = pd.read_csv("..\\dataset\\ebay_laptop_data_phase3.csv")
data
```

```python
feature = data[["Brand"]]
# feature["Brand_encoded"] = 
num_uniques = feature["Brand"].nunique()
feature_vals = feature["Brand"].unique()
code_list = [i + 1 for i in range(num_uniques)]

encoding_dict = {}
for value, code in zip(feature_vals, code_list):
    encoding_dict[value] = code 

feature["Encoding_brand"] = feature["Brand"].map(encoding_dict)
feature
```

```python
feature_name = data[["Brand"]].columns.tolist()
feature_name
```

```python
class LabelEncoder:
    def transform(feature: pd.DataFrame) -> pd.DataFrame: 
        # Define constants
        feature_name = feature_name.columns.tolist()
        num_uniques = feature[feature_name].nunique()
        feature_vals = feature[feature_name].unique()
        code_list = [i + 1 for i in range(num_uniques)]

        # Create encoding dictionary 
        encoding_dict = {}
        for value, code in zip(feature_vals, code_list):
            encoding_dict[value] = code 

        # map the feature value with its encoding value
        encoded_feature = feature["Brand"].map(encoding_dict)
        return encoded_feature
```

```python
label_encoder = LabelEncoder()
label_encoder.transform(data[["Brand"]])
```
